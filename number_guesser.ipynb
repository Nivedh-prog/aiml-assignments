```python
import random
num = random.randint(1,101)
guesses = 0
correct_guess = False

while(not correct_guess):
  print("Welcome to number game")
  inp_guess = int(input("Enter a number between 1 and 100 : "))
  guesses+=1

  if inp_guess == num:
    print("Congrats you found the number",num," in ",guesses," attempts")
    replay = input("Want to play again? y/n : ")
    if replay == 'y':
      print("Welcome Again to number game")
      num = random.randint(1,101)
      guesses = 0
    elif replay=='n':
      correct_guess = True
  elif num>inp_guess:
    print("Sorry too low")
  elif num<inp_guess:
    print("Sorry too high")


```
